```python
from bokeh.plotting import figure, show

# prepare some data
x = trips.index
y1 = trips['tripduration']
y2 = trips['number_of_trips']

# create a new plot with a title and axis labels
p = figure(title="Simple line example", x_axis_label="x", y_axis_label="y")

# add a line renderer with legend and line thickness
p.line(x, y1, legend_label="tripduration", color="blue", line_width=2)
p.line(x, y2, legend_label="number_of_trips", color="red", line_width=2)

# show the results
show(p)
```
