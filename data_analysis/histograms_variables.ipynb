```python
import pandas as pd
from modules.DataPreparation import DataPreparation
data_folder = '../data/'

def load_trips_time_series(year):
    data_folder = '../data/'
    trips = pd.read_csv(data_folder + 'trips/loaded_trips/trips_'+year + '.csv')
    dp = DataPreparation()
    trips = dp.transform_to_datetime(trips, ['starttime', 'stoptime'])
    trips = dp.transform_to_time_series(trips, 'starttime')
    return trips

def trips_to_time_series(trips):
    dp = DataPreparation()
    trips = dp.transform_to_datetime(trips, ['starttime', 'stoptime'])
    trips = dp.transform_to_time_series(trips, 'starttime')
    return trips
```

```python
trips_2018 = load_trips_time_series('2018')
trips_2019 = load_trips_time_series('2019')
```

```python
trips_2020 = load_trips_time_series('2020')
```

```python
trips_2021 = load_trips_time_series('2021')
```

```python
trips_2022 = load_trips_time_series('2022')
```

```python
trips_2020_antes_covid = trips_2020[:'2020-3-05']
trips_2020_depois_covid = trips_2020['2020-3-06':]
```

```python
data_folder = '../data/'
```

```python
trips_2020_antes_covid.to_csv(data_folder + 'trips/loaded_trips/trips_2020_before_covid.csv')
```

```python
trips_2020_depois_covid.to_csv(data_folder + 'trips/loaded_trips/trips_2020_after_covid.csv')
```

```python
trips_before_covid = trips_2018.append(trips_2019)
trips_before_covid = trips_before_covid.append(trips_2020_antes_covid)
```

```python
trips_after_covid = trips_2020_depois_covid.append(trips_2021)
trips_after_covid = trips_after_covid.append(trips_2022)
```

```python
trips_before_covid
```

```python
trips_after_covid
```

```python
trips = trips_before_covid
```

```python
trips = trips_after_covid
```

```python
trips_longer_4 = trips['tripduration'][trips['tripduration'] >= 14400].count()
```

```python
trips_dropped = trips_longer_4 / trips['tripduration'].count()
trips_dropped
```

```python
trips = trips['tripduration'][trips['tripduration'] < 14400]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from  matplotlib.ticker import FuncFormatter
ax = sns.histplot(data=trips, stat='percent', binwidth=1200)
for i in ax.containers:
    ax.bar_label(i,fmt='%.1f')
ax.set_xticks(range(0,14401,1200))
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: x//120))
ax.set(xlabel='Trip duration (min)', ylabel='Percent of trips (%)')
ax.set(title='Trip duration distribution before COVID-19 (< than 4h) \n Trips dropped = {number:.3f}%'.format(number = trips_dropped))
plt.savefig(data_folder + 'charts/histograms/trip_duration_before_covid.png')
```

```python
from  matplotlib.ticker import FuncFormatter
ax = sns.histplot(data=trips, stat='percent', binwidth=1200, kde=True)
ax.set_xticks(range(0,14401,1200))
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: x//120))
ax.set(xlabel='Trip duration (min)', ylabel='Percent of trips (%)')
ax.set(title='Trip duration distribution before COVID-19 (< than 4h) \n Trips dropped = {number:.3f}%'.format(number = trips_dropped))
plt.savefig(data_folder + 'charts/histograms/trip_duration_before_covid_kde.png')
```
