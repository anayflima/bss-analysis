```python
import pandas as pd
from modules.DataPreparation import DataPreparation
data_folder = '../data/'

def load_trips_time_series(year):
    data_folder = '../data/'
    trips = pd.read_csv(data_folder + 'trips/loaded_trips/trips_'+year + '.csv')
    dp = DataPreparation()
    trips = dp.transform_to_datetime(trips, ['starttime', 'stoptime'])
    trips = dp.transform_to_time_series(trips, 'starttime')
    return trips

def trips_to_time_series(trips):
    dp = DataPreparation()
    trips = dp.transform_to_datetime(trips, ['starttime', 'stoptime'])
    trips = dp.transform_to_time_series(trips, 'starttime')
    return trips
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from  matplotlib.ticker import FuncFormatter
fig, ax = plt.subplots()
fig.set_figwidth(8)
fig.set_figheight(6)

ax = sns.histplot(data=trips, stat='percent', binwidth=1200)
for i in ax.containers:
    ax.bar_label(i,fmt='%.1f', fontsize=13)
ax.set_xticks(range(0,14401,1200))
ax.set_ylim(bottom = 0, top = 70)
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: x//60))

ax.set_ylabel('Percent of trips (%)', color='k', fontsize=15, labelpad = 10)
ax.set_xlabel('Trip duration (min)', color='k', fontsize=15, labelpad = 10)

for label in (ax.get_xticklabels() + ax.get_yticklabels()):
        label.set_fontsize(12)

plt.title('Before pandemic', size=18, pad = 10)

# ax.set(xlabel='Trip duration (min)', ylabel='Percent of trips (%)', fontsize=22)
# ax.set(title='Trip duration distribution before COVID-19 (< than 4h) \n Trips dropped = {number:.3f}%'.format(number = trips_dropped))
# ax.set(title='Before pandemic')
# ax.title.set_fontsize(18)
# plt.savefig(data_folder + 'charts/histograms/trip_duration_before_covid.png')
```
