```python
import pandas as pd
```

```python
data_folder = '../data/'
year = '2021'
trips = pd.read_csv(data_folder + 'trips/loaded_trips/trips_'+year+'.csv')

from modules.DataPreparation import DataPreparation
dp = DataPreparation()
trips = dp.transform_to_datetime(trips, ['starttime', 'stoptime'])
trips = dp.transform_to_time_series(trips, 'starttime')
trips
```

```python
destination_folder = data_folder + 'trips/analysis/trips_by_month/'

for month in range(1,13,1):
    print(year+'-'+str(month)+':'+year+'-'+ str(month))
    trips_month = trips[year+'-'+str(month):year+'-'+ str(month)]
    filename = year + '_' + str(month)+ '.csv'
    trips_month.to_csv(destination_folder + filename)
```
