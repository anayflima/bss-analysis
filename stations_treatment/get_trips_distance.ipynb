```python
import pandas as pd

stations_distances = pd.read_csv('../data/stations/distance/stations_distance.csv')

stations_distances
```

```python
year = '2020'
trips = pd.read_csv("../data/trips/loaded_trips/trips_" +year+ ".csv")
trips
```

```python
trips = trips.drop(['id_y'], axis = 1)
```

```python
merge = trips.drop(['Unnamed: 0'], axis = 1)
```

```python
merge
```

```python
len(merge.columns)
```

```python
trips = trips.drop(['initial_station_latitude', 'initial_station_longitude',
                    'final_station_latitude', 'final_station_longitude'], axis = 1)
```

```python
merge = trips.merge(stations_distances, how = 'left', left_on=['index_start', 'index_end'],
                    right_on=['id_x', 'id_y'])

merge = merge.drop(['id_x', 'id_y'], axis = 1)
merge
```

```python
merge.to_csv("../data/trips/loaded_trips/trips_" +year+ ".csv", index=False)
```
