```python
import pandas as pd
data_folder = '../data/'
```

```python
year = '2021'
trips = pd.read_csv(data_folder + 'trips/loaded_trips/trips_' + year + '.csv')
trips
```

```python
trips = trips.rename(columns={'distance': 'distance_in_meters_bicycle'})
trips
```

```python
trips['distance_in_meters_bicycle'] = trips['distance_in_meters_bicycle'].apply(lambda x : x/1000)
```

```python
trips
```

```python
trips = trips.rename(columns={'index_start': 'start station id',
                             'start_station_name' : 'start station name',
                             'end_station_name' : 'end station name',
                             'index_end' : 'end station id'})
trips
```

```python
trips = trips.rename(columns={'distance_in_meters_bicycle': 'distance',
                             'start_station_id' : 'index_start',
                             'end_station_id' : 'index_end'})
trips
```

```python
trips.to_csv(data_folder + 'trips/loaded_trips/trips_' + year + '.csv', index=False)
```

```python
stations = pd.read_csv(data_folder + 'stations/distance/stations_distance.csv')
stations
```

```python
stations = stations.rename(columns={'id_source': 'id_x',
                             'id_destination' : 'id_y',
                             'distance_in_meters_bicycle' : 'distance'})
stations
```

```python
stations['distance'] = stations['distance'].apply(lambda x : x/1000)
```

```python
stations
```

```python
stations.to_csv(data_folder + 'stations/distance/stations_distance.csv', index=False)
```
