```python
year = '2022'
trips = pd.read_csv(data_folder + 'trips/loaded_trips/trips_' + year + '.csv')
trips
```

```python
try:
    trips = trips.drop(['distance_in_meters_bicycle_x'], axis = 1)
    trips = trips.rename(columns={'distance_in_meters_bicycle_y': 'distance_in_meters_bicycle'})
except:
    print('no distance_in_meters_bicycle_x')
```

```python
merge = trips.merge(stations_name, how = 'left', left_on=['start_station_id'],
                    right_on=['id'])

merge = merge.drop(['id'], axis = 1)

merge = merge.rename(columns={'start_station_name': 'start_station_name_old',
                                'name': 'start_station_name'})

merge = merge.merge(stations_name, how = 'left', left_on=['end_station_id'],
                    right_on=['id'])

merge = merge.drop(['id'], axis = 1)

merge = merge.rename(columns={'end_station_name': 'end_station_name_old',
                                'name': 'end_station_name'})

merge
```

```python
merge.to_csv(data_folder + 'trips/loaded_trips/trips_' + year + '.csv', index=False)
```
