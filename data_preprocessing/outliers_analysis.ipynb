# Outlier analysis

## Trip duration

```python
import pandas as pd
```

```python
trips = pd.read_csv('../data/trips/loaded_trips/trips_2018.csv')
import sys
sys.path.append('../data_analysis')
from modules.DataPreparation import DataPreparation
dp = DataPreparation()
trips = dp.transform_to_datetime(trips, ['starttime', 'stoptime'])
trips = dp.transform_to_time_series(trips, 'starttime')
```

We'll consider as outliers all trip durations greater than 12 hours (12*3600 seconds). Therefore, let's replace this column with null values when the value is greater than 43200. 

```python
len(trips[trips['tripduration'] > 12 * 3600])
```

```python
trips_selection = trips['2018-12']
len(trips_selection[trips_selection['tripduration'] > 12 * 3600])
```

```python
import numpy as np
trips_treated = trips.copy(deep=True)
trips_treated['tripduration'] = np.where(trips_treated['tripduration'] > 12*3600,
                                            np.nan, trips_treated['tripduration'])
trips_treated
```

```python
print(trips['tripduration'].isna().sum())
print(trips_treated['tripduration'].isna().sum())
```
